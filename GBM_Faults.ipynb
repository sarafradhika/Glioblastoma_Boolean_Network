```python
def Fault_to_Metric(mutations,types,N):
    default_nodes = np.zeros(N)
    default_faults = [False for i in range(N)] 
    
    #Assigning truth values to selected gene   
    for i in range(0,len(mutations)):
        default_nodes[mutations[i]] = 1
        default_faults[mutations[i]] = types[i]
    
    proApoptosis,antiApoptosis,proArrest,antiArrest,proAngiogenesis,antiAngiogenesis = BN(default_nodes,default_faults)
    CD = np.true_divide(proApoptosis,antiApoptosis) #CD is apoptosis
    ARREST = np.true_divide(proArrest,antiArrest)
    ANGIO = np.true_divide(antiAngiogenesis,proAngiogenesis)
    ConvexSum = np.average([CD,ARREST],weights=[14/20,6/20])
    
    metric = pd.Series({'proApoptosis':proApoptosis,'antiApoptosis':antiApoptosis,'proArrest':proArrest,'antiArrest':antiArrest,'proAngiogenesis':proAngiogenesis,'antiAngiogenesis':antiAngiogenesis,'ConvexSum': ConvexSum,'ApopRatio':CD,'ArrestRatio':ARREST,'AngioRatio':ANGIO})
    
    return metric
```

```python
def Drug_to_Metric(THISmutations,THIStypes,targets,actions,N):
    
    mutated_nodes = np.zeros(N)
    mutated_faults = [False for i in range(N)] 

    for i in range(0,len(THISmutations)):
        mutated_nodes[THISmutations[i]] = 1
        mutated_faults[THISmutations[i]] = THIStypes[i]
    
    #Assigning truth values to selected gene   
    for i in range(0,len(targets)):
        mutated_nodes[targets[i]] = 1
        mutated_faults[targets[i]] = actions[i]
    
    proApoptosis,antiApoptosis,proArrest,antiArrest,proAngiogenesis,antiAngiogenesis = BN(mutated_nodes,mutated_faults)
    CD = np.true_divide(proApoptosis,antiApoptosis) #CD is mutated_nodes
    ARREST = np.true_divide(proArrest,antiArrest)
    ANGIO = np.true_divide(antiAngiogenesis,proAngiogenesis)
    ConvexSum = np.average([CD,ARREST],weights=[14/20,6/20])
    
    metric = pd.Series({'proApoptosis':proApoptosis,'antiApoptosis':antiApoptosis,'proArrest':proArrest,'antiArrest':antiArrest,'proAngiogenesis':proAngiogenesis,'antiAngiogenesis':antiAngiogenesis,'ConvexSum': ConvexSum,'ApopRatio':CD,'ArrestRatio':ARREST,'AngioRatio':ANGIO})
    
    
    return metric
```
